```python
import random
import matplotlib.pyplot as plt

random.seed(42)

class Data:
    def __init__(self, pts: int):
        self.pts = pts

    def points(self):
        xs = []
        for i in range(self.pts):
            xs.append([random.random() - 0.5, random.random() - 0.5])
        return xs

    def diagonal(self):
        xs = self.points()
        ys = []
        for x in xs:
            if x[0] >= x[1]:
                ys.append(1)
            else:
                ys.append(0)
        return xs, ys


def plot_diagonal():

    data = Data(200)
    X, Y = data.diagonal()
    X0 = [x for x, y in zip(X, Y) if y == 0]
    X1 = [x for x, y in zip(X, Y) if y == 1]

    plt.figure(figsize=(4, 4))
    plt.scatter([x[0] for x in X0], [x[1] for x in X0], color='blue', label='Class 0')
    plt.scatter([x[0] for x in X1], [x[1] for x in X1], color='red', label='Class 1')


    plt.plot([-0.5, 0.5], [-0.5, 0.5], 'k--', label='y = x')

    plt.xlabel('x_1')
    plt.ylabel('x_2')
    plt.title('Diagonal')
    plt.legend(bbox_to_anchor=(1, 1))
    plt.show()
plot_diagonal()
```

```python
stride = 3
padding = 3
dilation = 3

x = Tensor(np.arange(100000).tolist(), (64, 3, 5, 5), history=History())
y = Tensor(np.arange(100000).tolist(), (4, 3, 2, 2), history=History())
out = x.conv2d(y, stride, padding, dilation)
print(out.shape)
out.backward()
print(x.gradient.shape)
print(y.gradient.shape)
print(x.gradient)
print(y.gradient)
```

```python
# Example input: (batch, in_channels, height, width)
a = torch.tensor(np.arange(64 * 3 * 5 * 5, dtype=float).tolist(), requires_grad=True).reshape(64, 3, 5, 5)
a.retain_grad()
# Example weight: (out_channels, in_channels, kernel_height, kernel_width)
b = torch.tensor(np.arange(4 * 3 * 2 * 2, dtype=float).tolist(), requires_grad=True).reshape(4, 3, 2, 2)
b.retain_grad()
# Apply convolution (no bias, stride=1, padding=0)
c = F.conv2d(a, b, stride=stride, padding=padding, dilation=dilation)
print(c.shape)
c.sum().backward()
print(a.grad.shape)
print(b.grad.shape)
print(a.grad)
print(b.grad)
```

```python
(B, C_in, H, W) = (64, 1, 28, 28)
(C_out, _, H_k, W_k) = (3, 3, 28, 28)

I_padded = np.arange(B * C_in * H * W).reshape(B, C_in, H, W)
K_dilated = np.arange(C_out * C_in * H_k * W_k).reshape(C_out, C_in, H_k, W_k)
(S_h, S_w) = (1, 1)
H_out = (H - H_k) // S_h + 1
W_out = (W - W_k) // S_w + 1



output = np.arange(B * C_out * H_out * W_out).reshape(B, C_out, H_out, W_out)
for b in range(B):
    for c_out in range(C_out):
        for i in range(H_out):
            for j in range(W_out):
                s = 0
                for c_in in range(C_in):
                    for p in range(H_k):
                        for q in range(W_k):
                            s += I_padded[b, c_in, i * S_h + p, j * S_w + q] * K_dilated[c_out, c_in, p, q]
                output[b, c_out, i, j] = s
```
